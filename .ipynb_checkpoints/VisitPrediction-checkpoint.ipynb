```python
import pandas as pd
import numpy as np
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df_load = pd.read_csv('data.jabarprov.go.id/disparbud-od_15367_jml_pengunjung_ke_objek_wisata__jenis_wisatawan_data.csv')
df_load.rename({'nama_kabupaten_kota': 'KabupatenKota', 'jenis_wisatawan': 'Jenis','jumlah_pengunjung': 'Total','tahun': 'Tahun'}, axis=1,inplace=True)
df = df_load[['KabupatenKota','Jenis','Total','Tahun']]
df.head(350)
```

```python
# count rows
df.shape
```

```python
df.groupby(['KabupatenKota','Jenis','Tahun']).sum()
```

```python
df_years = df.loc[:,'Tahun']
df_sum = df.loc[:,'Total']
```

```python
print(df_years)
print(df_sum)
```

```python
def visitPredicts(df_years,df_sum, x):
    #Create and train an SVR model using a linear kernel
    lin_svr = SVR(kernel='linear', C=1e3)
    poly_svr = SVR(kernel='poly', C=1e3, degree=7)
    rbf_svr = SVR(kernel='rbf', C=1e3, gamma=0.00193)
    
    lin_svr.fit(df_years, df_sum)
    poly_svr.fit(df_years, df_sum)#Create and train an SVR model using a RBF kernel
    rbf_svr.fit(df_years, df_sum)
    
    LnReg = LinearRegression()
    LnReg.fit(df_years,df_sum)
    
    #Plot the models on a graph to see which has the best fit
    plt.figure(figsize=(16,8))
    plt.scatter(df_years, df_sum, color = 'black', label='Original Data')
    plt.plot(df_years, rbf_svr.predict(df_years), color = 'green', label='RBF Model')
    plt.plot(df_years, poly_svr.predict(df_years), color = 'orange', label='Polynomial Model')
    plt.plot(df_years, lin_svr.predict(df_years), color = 'purple', label='Linear Model')
    plt.xlabel('Tahun')
    plt.ylabel('Prediksi Terakhir')
    plt.title('Support Vector Regression')
    plt.legend()
    plt.show()
    
    return rbf_svr.predict(x)[0],lin_svr.predict(x)[0],poly_svr.predict(x)[0],LnReg.predict(x)[0]
```
